## Import Modules

Let's start with importing modules. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generating Data

Let's generate data for Linear Regression. We'll do this using `np.random.normal(number_of_rows, number_of_features)` which generates data using **Normal Distribution** (a.k.a Gaussian Distribution) over `[0, 1)`.

```python
# generate data
# data = np.random.rand(1000,2) # (x, y)
mu = 0.0
su = 1.0 
data = np.random.normal(mu, su, 1000)
labels = np.random.rand(1000, 1)

# print("Data Generated of shape: {} and length: {}".format(data.shape, len(data)))

# # train test split: 70-30 ratio
X_train, X_test, y_train, y_test = train_test_split(data, labels[:, 0], test_size=0.3)
```

## Plot Data

Scatter Plot is usually the best plot for Linear Regression. It gives us a nice idea about the data.

```python
plt.scatter(data, labels) # plot the data using scatter plot
plt.show()
```

## Data Transformation

`sklearn.linear_model.LinearRegression.fit()` method expects a 2D array, and thus we'll have to reshape it to this format:

<center>`[[0.08], [0.84], ..., [0.9]]`</center>

```python
# sample data format
print("Training sample: {}\nTesting sample: {}\n".format(X_train[:5], y_train[:5]))
```

```python
# reshape data to 2D array
X_train_reshaped = X_train.reshape(-1, 1)
y_train_reshaped = y_train.reshape(-1, 1)
X_test_reshaped = X_test.reshape(-1, 1)
y_test_reshaped = y_test.reshape(-1, 1)

print(X_train_reshaped.shape, y_train_reshaped.shape)
```

## Training

```python
# initialize and fit model to the data
model = LinearRegression().fit(X_train_reshaped, y_train_reshaped)
print("Parameters of Regression Line")
print("Slope: %.2f, Intercept: %.2f" % (model.coef_[0], model.intercept_[0]))
```

## Testing

```python
# make predictions
predictions = model.predict(X_test_reshaped)

print("Mean Squared Error: %.2f" % np.sum(predictions - y_test_reshaped)**2)
```

## Plot Results

```python
plt.scatter(X_test_reshaped, y_test_reshaped, label='Original Dataset')
plt.scatter(X_test_reshaped, predictions, c = 'g', label='Predictions')
plt.xlabel('X axis (feature)')
plt.ylabel('Y axis (target)')
plt.title('Simple Linear Regression Results')
plt.legend()
```
